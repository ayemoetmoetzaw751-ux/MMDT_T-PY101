# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_quantity= int(input("Enter your quantity of apples: "))
apple_price = float(input("Price of an apple: "))

orange_quantity = int(input("Enter your quantity of oranges: "))
orange_price = float(input("Price of an orange: "))

mango_quantity = int(input("Enter your quantity of mango: "))
mango_price = float(input("Price of a mango: "))

cost_apple = apple_price * apple_quantity
cost_orange = orange_price * orange_quantity
cost_mango = mango_price * mango_quantity

total_cost = cost_apple + cost_orange + cost_mango
print("Yourtotal_cost is {}".format(total_cost))
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
total_seconds = int(input("Enter the total seconds: "))
minutes = total_seconds // 60
remaining_sec = total_seconds % 60
print("{} seconds is equal to {} minutes and {} seconds.".format(total_seconds, minutes, remaining_sec))
```

# Ex 10: Using Math Module

```python
import math
num = float(input("Enter a number: "))
sin_value = math.sin(num)
cosine_value = math.cos(num)
tangent_value = math.tan(num)
print("Your {} has those {} sine value, {} cosine value, and {} tangent value".format(num, sin_value, cosine_value, tangent_value))
```
